# Importing Libraries

```python
%matplotlib inline
from shapely.geometry import Point
from geopandas import datasets, GeoDataFrame, read_file
from geopandas.tools import overlay
```

# Reading dataset

```python
zippath = datasets.get_path('nybb')
polydf = read_file(zippath)
```

# Creating circles

```python
b = [int(x) for x in polydf.total_bounds]
N = 10
polydf2 = GeoDataFrame([
    {'geometry': Point(x, y).buffer(10000), 'value1': x + y, 'value2': x - y}
    for x, y in zip(range(b[0], b[2], int((b[2] - b[0]) / N)),
                    range(b[1], b[3], int((b[3] - b[1]) / N)))])
```

# Plotting df

```python
polydf.plot(figsize=(8,8), cmap = 'tab20b', legend=True)
```

## Plotting GeoDataFrame having a sequentially generated set of circles in the same geographic space.

```python
polydf2.plot(cmap='tab20b')
```

# Using overlay function (how = "intersection")

```python
newdf = polydf.overlay(polydf2, how="intersection")
newdf.plot(cmap='tab20b', figsize = (5,5))
```

# Using overlay function (how = "union")

```python
newdf = polydf.overlay(polydf2, how="union")
newdf.plot(cmap='tab20b', figsize = (5,5))
```

# Using overlay function (how = "identity")

```python
newdf = polydf.overlay(polydf2, how="identity")
newdf.plot(cmap='tab20b', figsize = (5,5))
```

# Using overlay function (how = "symmetric_difference")

```python
newdf = polydf.overlay(polydf2, how="symmetric_difference")
newdf.plot(cmap='tab20b', figsize = (5,5))
```

# Using overlay function (how = "difference")

```python
newdf = polydf.overlay(polydf2, how="difference")
newdf.plot(cmap='tab20b', figsize = (5,5))
```
